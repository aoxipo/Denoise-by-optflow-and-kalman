```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def decode_file(file_path):
    print("path :", file_path)
    f = open( file_path ,"r")
    context = f.read() 
    iter_list = context.split(">1")
    total = len(iter_list)
    x_list = []
    y_list = []
    for node in iter_list:
        if( len(node) < 10):
            continue
        landmark_total = node.split(',')
        sum_x = 0
        sum_y = 0
        landmark_number = len(landmark_total)
        for i_ in landmark_total:
            if(len(i_)>10):
                sum_x += int(float( i_.split("\n")[1].split(":")[1]) * 640)
                sum_y += int(float( i_.split("\n")[2].split(":")[1]) * 480)
            else:
                landmark_number -= 1
        x_list.append(sum_x/landmark_number)
        y_list.append(sum_y/landmark_number)
    return np.array(x_list), np.array(y_list)
```

```python
f_root = "./save/"
f_origin = "origin_result_face.txt"
f_optflow = "optflow_result_face.txt"
f_lin_fitting = "lin_fitting_result_face.txt"
f_kalman = "kalman_result_face.txt"
save_path_x = "face_x.png"
save_path_y = "face_y.png"
```

```python
x_origin_list, y_origin_list = decode_file(f_root+f_origin)
x_f_optflow_list, y_f_optflow_list = decode_file(f_root+f_optflow)
x_f_lin_fitting_list, y_f_lin_fitting_list = decode_file(f_root+f_lin_fitting)
x_f_kalman_list, y_f_kalman_list = decode_file(f_root+f_kalman)
```

```python
def png_plot(save_x, save_y ,x_origin_list, x_f_optflow_list, x_f_lin_fitting_list, x_f_kalman_list, y_origin_list, y_f_optflow_list, y_f_lin_fitting_list, y_f_kalman_list):

    start = 100

    plt.plot(x_origin_list[start:] -  x_f_optflow_list[start:], color='green', label='origin')
    plt.plot(x_f_optflow_list[start:] - x_f_optflow_list[start:], color='red', label='optflow')
    plt.plot(x_f_lin_fitting_list[start:] - x_f_optflow_list[start:], color='blue', label='line fitting')
    plt.plot(x_f_kalman_list[start:] - x_f_optflow_list[start:], color='yellow', label='kalman')
    plt.xlabel("x axis")
    plt.legend()
    plt.savefig(save_x, dpi=600, bbox_inches='tight')
    plt.show()

    plt.plot( y_origin_list[start:] - y_f_optflow_list[start:], color='green', label='origin')
    plt.plot( y_f_optflow_list[start:] - y_f_optflow_list[start:], color='red', label='optflow')
    plt.plot( y_f_lin_fitting_list[start:] - y_f_optflow_list[start:] - 176.4-65, color='blue', label='line fitting')
    plt.plot( y_f_kalman_list[start:] - y_f_optflow_list[start:], color='yellow', label='kalman')
    plt.xlabel("y axis")
    plt.legend()
    plt.savefig(save_y, dpi=600, bbox_inches='tight')
    plt.show()
```

```python
png_plot(f_root+save_path_x, f_root+save_path_y, x_origin_list, x_f_optflow_list, x_f_lin_fitting_list, x_f_kalman_list, y_origin_list, y_f_optflow_list, y_f_lin_fitting_list, y_f_kalman_list)
```
